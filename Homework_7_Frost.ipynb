```python
#importing relevant modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
#importing data as numpy arrays
url1 = 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/acc_hits_hw5.csv'
acc_hits = np.loadtxt(url1, delimiter = ',')

url2 = 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/acc_fa_hw5.csv'
acc_fa = np.loadtxt(url2, delimiter = ',')

```

```python
#printing the shape of each array
print(acc_hits.shape)
print(acc_fa.shape)
```

```python
#creating a time variable
time = np.linspace(0,4, acc_hits.shape[1])
```

```python
#printing the sampling rate
sampling_rate = 8001/4

print(sampling_rate)
```

```python
#Creating two subplots and plotting time on the x axis and the sample of interest on the y

fig, ax = plt.subplots(1,2,figsize=(12,4))

ax[0].plot(time, acc_hits[0])
ax[0].set_xlabel('Time')
ax[0].set_ylabel('LFP(mV)')
ax[0].set_title('First Trial - Hits')


ax[1].plot(time, acc_fa[0])
ax[1].set_xlabel('Time')
ax[1].set_ylabel('LFP (mV)')
ax[1].set_title('First Trial - False Alarms')
```

I think that the dominant oscillation is at about 4-6 hz. The frequency differs between the two trials visually, with the hits having a distinct path and the false alarms seeming to not have as distinct a pattern.

```python
#using the mean hits to create ERP hits and plotting it
#creating axv and axh lines to display where touch occurs

erp_hits = np.mean(acc_hits, axis=0)

plt.figure(figsize=(8, 4))
plt.plot(time, erp_hits, color='blue', label='ERP - Hits')
plt.axvline(x=2, color='black', linewidth=1.5, label='Screen touch')
plt.axhline(y=0, color='black', linewidth=1.5,)
plt.xlabel('Time [s]')
plt.ylabel('LFP (mV)')
plt.title('ERP - Hits')
plt.legend()
plt.show()
```

Yes we see consistent activity in the LFP surrounding the screen touch. Immediately after the screen touch you can observe the difference in LFP that occurs.

```python
#Using the means to create erp_fa and plotting it on top of the previous code

erp_hits = np.mean(acc_hits, axis=0)
erp_fa = np.mean(acc_fa, axis=0)

plt.figure(figsize=(8, 4))
plt.plot(time, erp_hits, color='blue', label='ERP - Hits')
plt.plot(time, erp_fa, color='red', label='ERP - False Alarms')
plt.axvline(x=2, color='black', linewidth=1.5, label='Screen touch')
plt.axhline(y=0, color='black', linewidth=1.5, linestyle='--')
plt.xlabel('Time [s]')
plt.ylabel('LFP (mV)')
plt.title('ERP - Hits vs. False Alarms')
plt.legend()
plt.show()
```

Yes, there are, the false alarms don't show any significant change in LFP
