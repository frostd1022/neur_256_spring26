```python
#Importing numpy, matplotlib.pyplot and statistics

import numpy as np

import matplotlib.pyplot as plt

import statistics as st
```

```python
#importing v_rest1 and v_rest 2 as our data in the Github

url = 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_pvalb.csv'
v_rest1 = np.loadtxt(url)

url = 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_sst.csv'
v_rest2 = np.loadtxt(url)

```

```python
#concatonating v_rest 1 and v_rest2 into v_rest

v_rest = np.vstack((v_rest1 + v_rest2))
```

```python
#Checking the size of v_rest

print(v_rest.size)
```

```python
#First row of v_rest

v_rest[0]
```

```python
#Returns the sum of the values in the 3rd column of v_rest

v_rest[2].sum()
```

```python
#Returns second-to-last value in second row of v_rest

v_rest[-2]
```

```python
#cell that creates a histogram of the values in the first row of v_rest

plt.hist(v_rest[0])
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of First Row')
```

```python
#Gets the mean of v_rest and assigns the rows to different variables

v_rest_mean_sst = st.mean(v_rest[0])
v_rest_mean_pvalb = st.mean(v_rest[1])

print(v_rest_mean_sst, v_rest_mean_pvalb)
```

The difference in means is about 20. The larger in means is the mean of the pvalb values.  

```python
#importing scipy.stats

import scipy.stats as stats
```

```python
#Conducting a t-test

stats.ttest_ind(v_rest1, v_rest2)
```

My pvalue is 7.15% or .0715. Since it is greater than 0.05, I would consider my data to be statistically insignificant.
