```python
#Importing data

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

```

```python
#Importing variables as arrays

url = 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_pvalb.csv'
v_rest_pvalb = np.loadtxt(url)

url = 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_sst.csv'
v_rest_sst = np.loadtxt(url)
```

```python


#finding the means of v_rest_sst and v_rest_pvalb
mean_sst = np.mean(v_rest_sst)
mean_pvalb = np.mean(v_rest_pvalb)

print("Mean SST:", mean_sst)
print("Mean Pvalb:", mean_pvalb)

#Performing t-test to test the null
t_stat , p_val = stats.ttest_ind(v_rest_sst, v_rest_pvalb)

print("T-test Results:" ,t_stat)
print("p-value:", p_val)

#Using an if statement to see if the p-value is greater than or less than .05
if p_val < 0.05:
  fig, ax = plt.subplots(1,2)

  #labeling axes
ax[0].hist(v_rest_sst)
ax[0].set_title("SST")

ax[1].hist(v_rest_pvalb)
ax[1].set_title("Pvalb")

ax[0].set_xlabel("Values")
ax[0].set_ylabel("Frequency")

ax[1].set_xlabel("Values")
ax[1].set_ylabel("Frequency")

plt.show()
```
