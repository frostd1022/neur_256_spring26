```python
#Importing relevant modules

import numpy as np
import matplotlib.pyplot as plt
```

```python
#Putting in relevant data

url1 = 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/spikes_left_trial1.csv'
url2 = 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/spikes_right_trial1.csv'
url3 = 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/spikes_left_trial2.csv'
url4 = 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/spikes_right_trial2.csv'

spikes_left_trial1 = np.loadtxt(url1, delimiter = ',')
spikes_right_trial1 = np.loadtxt(url2, delimiter = ',')
spikes_left_trial2 = np.loadtxt(url3, delimiter = ',')
spikes_right_trial2 = np.loadtxt(url4, delimiter = ',')
```

```python
#printing the length of each array

print(len(spikes_left_trial1))
print(len(spikes_right_trial1))
print(len(spikes_left_trial2))
print(len(spikes_right_trial2))
```

```python
# Calculating firing rates (spikes per second)
fr_left1 = len(spikes_left_trial1) / 30
fr_right1 = len(spikes_right_trial1) / 30
fr_left2 = len(spikes_left_trial2) / 30
fr_right2 = len(spikes_right_trial2) / 30

print(fr_left1, fr_right1, fr_left2, fr_right2)
```

I think it fires more before left trials, although it is pretty close between the two.

```python
#Made two subplots

fig, ax = plt.subplots(2, 1, figsize=(15,15))

#Spike train data from trial one in the first subplot

ax[0].plot(spikes_left_trial1, color = 'red')
ax[0].plot(spikes_right_trial1)
ax[0].set_ylabel('Spikes')
ax[0].set_xlabel('Time')
ax[0].set_title('Trial 1')

#Spike train data from trial 2 in the second subplot

ax[1].plot(spikes_left_trial2, color = 'red')
ax[1].plot(spikes_right_trial2)
ax[1].set_ylabel('Spikes')
ax[1].set_xlabel('Time')
ax[1].set_title('Trial 2')
```

I think that the neurons have more regular firing patterns when turning to the right and have more bursty firing patterns when firing to the left.

```python
#Defining the ISI values

isi_left1 = np.diff(spikes_left_trial1)
isi_right1 = np.diff(spikes_right_trial1)
isi_left2 = np.diff(spikes_left_trial2)
isi_right2 = np.diff(spikes_right_trial2)
```

```python
# Defining the bins
bins = np.arange(0, 2, 0.01)

# Plotting the histograms
fig, ax = plt.subplots(2, 2, figsize=(12,8))

ax[0,0].hist(isi_left1, bins=bins)
ax[0,0].set_title('Left Trial 1')

ax[0,1].hist(isi_right1, bins=bins)
ax[0,1].set_title('Right Trial 1')

ax[1,0].hist(isi_left2, bins=bins)
ax[1,0].set_title('Left Trial 2')

ax[1,1].hist(isi_right2, bins=bins)
ax[1,1].set_title('Right Trial 2')
```

It seems like the left trials have more burstiness while the right firing is more consistent

```python
# Time bins (0–30 sec, 1 sec bins)
bins = np.arange(0, 31, 1)

fr_left1_time, _ = np.histogram(spikes_left_trial1, bins=bins)
fr_right1_time, _ = np.histogram(spikes_right_trial1, bins=bins)
fr_left2_time, _ = np.histogram(spikes_left_trial2, bins=bins)
fr_right2_time, _ = np.histogram(spikes_right_trial2, bins=bins)

time = bins[:-1]

fig, ax = plt.subplots(2, 1, figsize=(10,6))

# Trial 1
ax[0].plot(time, fr_left1_time, color = 'red', label='Left')
ax[0].plot(time, fr_right1_time, color = 'blue', label='Right')
ax[0].set_title('Trial 1')
ax[0].set_ylabel('Firing Rate (Hz)')


# Trial 2
ax[1].plot(time, fr_left2_time, color = 'red', label='Left')
ax[1].plot(time, fr_right2_time, color = 'blue', label='Right')
ax[1].set_title('Trial 2')
ax[1].set_xlabel('Time (s)')
ax[1].set_ylabel('Firing Rate (Hz)')

```

I notice that at about 10 seconds for trial one and at about 20 seconds for trial two we see a pattern of consistent firing start for the left turning. The right however seems to be more spread out, aligning with the bursty firing pattern we saw previously.
